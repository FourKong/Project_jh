```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from mord import LogisticIT
from sklearn.metrics import accuracy_score
import pandas as pd
from google.colab import drive
import joblib
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_data.csv')
X = file.drop(['sat_all', 'pid'], axis=1)
y = file['sat_all']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
model = LogisticIT()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
coefficients = model.coef_
feature_names = X.columns

for i, coefficient in enumerate(coefficients):
    feature = feature_names[i]
    print(f"Feature: {feature}, Coefficient: {coefficient}")
```

```python
file = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_data.csv')
X = file.drop(['sat_all', 'sat_living', 'sat_health', 'sat_human','sat_achieve', 'pid'], axis=1)
datas = ['all', 'living', 'health', 'human']
df_result = pd.DataFrame()
for feature in X.columns :
  header = pd.DataFrame([f'{feature}'])
  df_result = pd.concat((df_result, header), ignore_index=True)
df_result = pd.concat((df_result, pd.DataFrame(['sat'])), ignore_index=True)
index=0
for a in datas:
    a_model = joblib.load(f'/content/drive/MyDrive/Colab Notebooks/{a}_model.pkl')
    df = pd.DataFrame({'feature': [], 'coef': []})

    for feature, coefficient in zip(X.columns, a_model.coef_):
        df2 = pd.DataFrame({'feature': [feature], 'coef': [coefficient]})
        df = pd.concat((df, df2), ignore_index=True)

    df3 = pd.DataFrame({'feature': ['sat'], 'coef': [index+10]})
    df = pd.concat((df, df3), ignore_index=True)
    df_result = pd.concat((df_result, df['coef']), ignore_index=True, axis=1, join='inner')
    index = index + 1

df_result = df_result.round(3)
print(df_result.T)
df_result.T.to_excel(excel_writer='coef_data.xlsx', index=False, header=False)
```
