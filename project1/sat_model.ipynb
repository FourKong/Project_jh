```python
import joblib
from sklearn.preprocessing import OneHotEncoder
from mord import LogisticIT
```

```python
def predict(userData):
  result = []
  all_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/all_model.pkl')
  living_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/living_model.pkl')
  health_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/health_model.pkl')
  human_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/human_model.pkl')
  f_data = userData.drop('sat_all',  axis=1)
  s_data = OneHotEncoder().fit_transform(f_data)
  t_data = s_data.reshape(1, -1)
  pre_data = all_model.predict(t_data)
  result.append(pre_data)
  for model in [living_model,health_model,human_model]:
    f_data = userData.drop(['sat_all', 'sat_living', 'sat_health', 'sat_human'],  axis=1)
    s_data = OneHotEncoder().fit_transform(f_data)
    t_data = s_data.reshape(1, -1)
    pre_data = model.predict(t_data)
    result.append(pre_data)
    return result
```
