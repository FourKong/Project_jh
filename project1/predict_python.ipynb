```python
import joblib
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
all_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/all_model.pkl')
file = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_data.csv')

X = file.drop(['sat_all', 'sat_achieve'], axis=1)

encoder = OneHotEncoder()
# 행 하나 추출의 경우 실제 실행 시엔 X. 받는 데이터가 1행임.
X_encoded = encoder.fit_transform(X)[99]
pre_data = X_encoded.reshape(1, -1)
predicted_sat_all = all_model.predict(pre_data)
print(predicted_sat_all)
```

```python
file = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final_data.csv')
X = file.drop(['sat_all', 'sat_living', 'sat_health', 'sat_human','sat_achieve'], axis=1)
datas = ['all', 'living', 'health', 'human']
df_result = pd.DataFrame()
for feature in X.columns :
  header = pd.DataFrame([f'{feature}'])
  df_result = pd.concat((df_result, header), ignore_index=True)
df_result = pd.concat((df_result, pd.DataFrame(['sat'])), ignore_index=True)
index=0
for a in datas:
    a_model = joblib.load(f'/content/drive/MyDrive/Colab Notebooks/{a}_model.pkl')
    df = pd.DataFrame({'feature': [], 'coef': []})

    for feature, coefficient in zip(X.columns, a_model.coef_):
        df2 = pd.DataFrame({'feature': [feature], 'coef': [coefficient]})
        df = pd.concat((df, df2), ignore_index=True)

    df3 = pd.DataFrame({'feature': ['sat'], 'coef': [index+10]})
    df = pd.concat((df, df3), ignore_index=True)
    df_result = pd.concat((df_result, df['coef']), ignore_index=True, axis=1, join='inner')
    index = index + 1
print(df_result.T)
df_result.T.to_excel(excel_writer='sample.xlsx', index=False, header=False)
```
