```python
from sklearn.linear_model import LogisticRegression
# house_rental 변수를 더미 변수로 변환
X_house_rental = pd.get_dummies(X['house_rental'], prefix='house_rental')

# 변환된 더미 변수와 다른 독립 변수를 합침
X_encoded = pd.concat([X.drop('house_rental', axis=1), X_house_rental], axis=1)

# 로지스틱 회귀 모델 적합
model = LogisticRegression()
model.fit(X_encoded, y)

# house_rental 변수의 각 수준별 계수 확인
coefficients = model.coef_[0][-5:]  # house_rental의 마지막 세 개 계수만 선택 (예시 기준, 실제 수준에 따라 조정 필요)
house_rental_levels = [1, 2, 3, 4, 5]  # 실제 수준에 따라 조정 필요

for level, coefficient in zip(house_rental_levels, coefficients):
    print(f"Level: {level}, Coefficient: {coefficient}")
```
